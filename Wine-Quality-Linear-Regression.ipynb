1. Using scikit-learn, then, build a regression model that predicts the quality of wine using any machine learning algorithms and report the training set error by sharing your Jupyter notebook. Your answer needs to obtain the RMSE (root-mean-square-error) score of at most 0.625. 

```python
# Import useful packages
%reset
import pandas as pd
import numpy as np
from sklearn import linear_model, preprocessing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Import, format, and explore dataset
wine = pd.read_csv("winequality-red.csv", sep = ';')
print(wine.keys())
print(wine.info())

col_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
x = pd.DataFrame(wine, columns = col_names)
print(x.head())

y = pd.DataFrame(wine, columns=['quality'])
print(y.head())

# Check for null values
print(x.isnull().sum())
```

```python
# Minmax normalization of data
x = x.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x_scaled = pd.DataFrame(x_scaled, columns=col_names)

# Create and view normalized dataframe
norm_dataset = pd.concat([x_scaled, y], axis=1, sort=False)
norm_dataset.head()
```

I first tried this assignment with the Lasso linear regression model because I wanted to have the ability to use feature selection. There are a lot of variables in this dataset, and I was worried that if all of them are taken into account, I may overfit my model. However, upon working with Lasso, I found that close to zero alpha values were performing better, and I recognized that simple linear regression may be a better option.

```python
# Prepping data for split
data = norm_dataset.loc[:,norm_dataset.columns!='quality']
result = norm_dataset['quality']

# Split training and testing data
x_train, x_test, y_train, y_test = train_test_split(data, result, random_state=42)

# Implementing linear regression
mymodel = LinearRegression(fit_intercept = True)
print(mymodel)

# Fit the model to training data
mymodel.fit(x_train, y_train)
print(mymodel.intercept_)
print(mymodel.coef_)
# Predict on testing data
predicted = mymodel.predict(x_test)

# Compute RMSE of predictions versus actual wine quality
mean = mean_squared_error(predicted, y_test) 
rmse = np.sqrt(mean)
print(rmse)
```

```python
# Use same code from earlier to import and format dataset
wine = pd.read_csv("winequality-red.csv", sep = ';')

col_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']
x = pd.DataFrame(wine, columns = col_names)

y = pd.DataFrame(wine, columns=['quality'])

# Check for null values
print(x.isnull().sum())

# Now we want to explore relationships to complete feature engineering for polynomial regression.
print(wine.corr())

# Generate polynomial features based on observed correlations
alc_power = wine['alcohol'].pow(40)
acid_power = wine['volatile acidity'].pow(15)
sulph_power = wine['sulphates'].pow(3)
ca_power = wine['citric acid'].pow(3)
so2_power = wine['total sulfur dioxide'].pow(10)
cl_power = wine['chlorides'].pow(9)
fix_acid = wine['fixed acidity'].pow(2)
dens = wine['density'].pow(2)
ph_pow = wine['pH'].pow(2)
alc_dens = wine['density']*wine['alcohol']
sug_acid = wine['residual sugar']*wine['free sulfur dioxide']
acid_alc = wine['citric acid']*wine['alcohol']
alc_chl = wine['alcohol']*wine['chlorides']*wine['citric acid']
acid_ph = wine['volatile acidity']*wine['pH']*wine['sulphates']
acid_sulph = wine['volatile acidity']*wine['sulphates']*wine['citric acid']*wine['pH']
acid_acid = wine['volatile acidity']*wine['citric acid']
citric_fix = wine['fixed acidity']*wine['citric acid']*wine['pH']
ca_ph = wine['citric acid']*wine['pH']
sulph_ca = wine['citric acid']*wine['sulphates']
sug_dens = wine['residual sugar']*wine['density']
tsd_alc = wine['total sulfur dioxide']*wine['alcohol']
dens_fix = wine['density']*wine['fixed acidity']*wine['alcohol']
sulph_sug = wine['sulphates']*wine['residual sugar']
x = pd.concat([x, alc_power, acid_power, sulph_power, ca_power, so2_power, cl_power, fix_acid, dens, ph_pow, alc_dens, sug_acid, acid_alc, alc_chl, acid_ph, acid_sulph, 
               acid_acid, citric_fix, ca_ph, sulph_ca, sug_dens, tsd_alc, dens_fix, sulph_sug], axis=1, sort=False)
```

```python
# Convert dataframes to numpy arrays to do calculations
y = y.to_numpy()
x = x.to_numpy()

# Formula for theta = (X^T * X)^-1 * X^T * y
# Implement using numpy operations to produce numpy array of coefficients
theta = np.linalg.inv(x.transpose().dot(x)).dot(x.transpose()).dot(y)
print(theta)
```

```python
# Multiply theta array by training data to generate predictions
yhat = np.dot(x, theta)
# Write predictions to original dataframe and explore visually
wine['pred'] = yhat.tolist()
print(wine)

# Compute RMSE of testing data using numpy operations
mean = np.square(np.subtract(y, yhat)).mean() 
rmse = np.sqrt(mean)
print(rmse)
```
